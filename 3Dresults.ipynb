```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML

fig = plt.figure()
ims = []

for i in range(1,7):
    result = np.load("../new/save/27/%s.npz" % i)
    initial = result['phi'][10]
    im = plt.imshow(initial, animated=True)
    ims.append([im])
    
anim = animation.ArtistAnimation(fig, ims, interval=100, blit=True,
                                repeat_delay=1000)
plt.show()
```
